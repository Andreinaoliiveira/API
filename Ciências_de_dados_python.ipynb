**TESTE T**

h0:  A média do uso de cartão de crédito para o grupo com tendência "baixa" é igual à média do grupo com tendência "alta"

h1: A média não é igual.

```python
import pandas as pd
from scipy import stats

grupo_baixa = dados[dados['Tendência'] == 'baixa']['Cartão de crédito']
grupo_alta = dados[dados['Tendência'] == 'alta']['Cartão de crédito']

# Realizando o teste T
t_stat, p_value = stats.ttest_ind(grupo_baixa, grupo_alta, equal_var=False)

print(f'Estatística t: {t_stat}')
print(f'Valor p: {p_value}')


alpha = 0.05  # nível de significância
if p_value < alpha:
    print("Rejeitamos a hipótese nula (H0): há uma diferença significativa entre os grupos.")
else:
    print("Não rejeitamos a hipótese nula (H0): não há diferença significativa entre os grupos.")
```

```python
import pandas as pd
import pingouin as pg

# Supondo que 'dados' já está definido e contém a variável 'Tendência'
# Convertendo a variável 'Tendência' em variáveis dummy
dados_dummies = pd.get_dummies(dados['Tendência'], drop_first=True)  # drop_first evita multicolinearidade

# Juntando as dummies ao DataFrame original
dados = pd.concat([dados, dados_dummies], axis=1)

# Realizando a regressão linear usando a variável dummy
# Supondo que as dummies têm os nomes 'media' e 'alta' após a conversão
# Use o nome da coluna correspondente na regressão
#aq
# Exemplo: se você tem 'alta' e 'media' como variáveis dummy
modelolin = pg.linear_regression(dados[['media', 'baixa']], dados['Cheque especial'])
print(modelolin)

```

Significância: Todos os coeficientes têm p-valores significativos, o que sugere que tanto a tendência média quanto a baixa têm um impacto significativo no uso do Cheque especial. O modelo tem um R² relativamente baixo, o que sugere que pode haver outras variáveis não incluídas que poderiam explicar mais variação no uso do Cheque especial.

```python
#cheque especial e crédito pessoal
modelolin = pg.linear_regression(dados[['Crédito pessoal']], dados['Cheque especial'])
print(modelolin)
```

**Interpretação:**
Com relação positiva linear moderada entre essas duas variáveis, temos:

No coeficiente do intercepto, em termos práticos, mesmo sem qualquer valor em crédito pessoal, ainda existe uma média de 0.036236 no cheque especial. O que faz sentido, uma vez que são meio de dívidas sem muita relação aparente.
